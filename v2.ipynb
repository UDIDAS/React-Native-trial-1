# Plots

```python
dataPr.columns
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.offline as pyo
```

```python
sort_indices = np.argsort(dataPr['processed'])
x = np.array(dataPr['processed'])[sort_indices]
y1 = np.array(dataPr['child'])[sort_indices]
y2 = np.array(dataPr['family_health'])[sort_indices]
y3 = np.array(dataPr['mandate'])[sort_indices]

# Create the first area chart
fig_AC1 = make_subplots(rows=3, cols=1)

# Add the first area chart to the first subplot
fig_AC1.add_trace(go.Scatter(x=x, y=y1, fill='tozeroy', name='child'), row=1, col=1)

# Add the second area chart to the second subplot
fig_AC1.add_trace(go.Scatter(x=x, y=y2, fill='tozeroy', name='family_health'), row=2, col=1)

# Add the third area chart to the third subplot
fig_AC1.add_trace(go.Scatter(x=x, y=y3, fill='tozeroy', name='mandate'), row=3, col=1)

# Update subplot layout
fig_AC1.update_layout(
    title='Area Charts',
    height=800
)

x_tick_labels = [val[:15]+'...' for val in x]

fig_AC1.update_xaxes(tickvals=[], ticktext=[], row=1, col=1)
fig_AC1.update_xaxes(tickvals=[], ticktext=[], row=2, col=1)
fig_AC1.update_xaxes(tickvals=x[::30], ticktext=x_tick_labels, row=3, col=1)

fig_AC1.show()
```

```python
pyo.plot(fig_AC1, filename='AreaChart1.html', auto_open=False)
```
